## Exercises

### Calculating $e^x$

Use recursion to compute the value of $e^x$ using the string

$$
e^x = \sum_{n=0}^{N-1} \frac{x^n}{n!}
$$

up to a fixed number $N$ of terms.

Now suppose you want to compute the series but instead of computing a fixed number of terms, you want to evaluate the function with a given relative error `epsilon`. If we define $e_N^x$ as the sum of the first $N$ elements of the series, the relative error at step $N$ is:

$$
\epsilon = \left| \frac{e_N^x - e_{N-1}^x}{e_{N-1}^x} \right|,
$$

Code a function that receives the value of $x$ and a value for the relative error $\epsilon$ and returns the value of $e^x$ such that the error in the calculation is less than $\epsilon$.

### Back to the _truco_

Retaking the [_truco_ exercise](_fsharp/14_Exercises.md):

The _envido_ in the _truco_ requires you to have two cards of the same suit, out of the three cards in your hand.
Assuming that you have 2 cards of the same suit, add their value + 20 (hence the importance of linking two cards of the same suit). The highest score of the _envido_ is 33 (7+6+20=33).
The figures or black cards (jack, horse and king) are worth 0 points for the bid. If you have 2 figures of the same suit, the bid score is 20.
Remember that in the _truco_, the eights and nines are not part of the deck.

- Write a function that, given three cards, determines if it is possible to form an _envido_.
- Write a function that, given two cards, returns the value of the _envido_, or some type of error if you do not have an _envido_ with said cards.
