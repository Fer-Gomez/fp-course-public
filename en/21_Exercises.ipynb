$$
e^x = \sum_{n=0}^{N-1} \frac{x^n}{n!}
$$

up to a fixed number $N$ of terms.

Now suppose you want to compute the series but instead of computing a fixed number of terms, you want to evaluate the function with a given relative error `epsilon`. modify
the above code to get this function.

### Back to the trick

Retaking the [trick exercise](_fsharp/14_Exercises.md):

The _raise_ in the trick requires you to have two cards of the same suit, out of the three cards in your hand.
Assuming that you have 2 cards of the same suit, add their value + 20 (hence the importance of linking two cards of the same suit). The highest score of the raise is 33 (7+6+20=33).
The figures or black cards (jack, horse and king) are worth 0 points for the bid. If you have 2 figures of the same suit, the bid score is 20.
Remember that in the trick, the eights and nines are not part of the deck.

- Write a function that, given three cards, determines if it is possible to form a raise.
- Write a function that, given two cards, returns the value of the raise, or some type of error if you do not have a raise with said cards.
