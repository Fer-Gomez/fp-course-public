## Input/Output

Perhaps one of the main tasks of all code is to read (and write) information in files. To do this, F# uses at its base the functions that the .NET ecosystem offers, [with numerous alternatives](). Let's see some of them:

```python
let fileName = "../data/El reloj de arena.txt"


let readFile (fileName: string) =  
    let lines = File.ReadAllLines(fileName)
    lines 
```

```python
let lines = readFile fileName 
```

The `ReadAllLines` function returns a sequence, where each line is an element:

```python
lines
|> Seq.iteri (fun i l -> printfn "%d: %s" i l)
```

```python
lines
|> Seq.where (fun l -> l.Contains("tiempo"))
|> Seq.iteri (fun i l -> printfn "%d: %s" i l)
```

Using some methods of `seq`, we can process the poem and separate it into stanzas:

```python
lines
|> Seq.skip 3
|> Seq.chunkBySize 5
|> Seq.map (fun estrofa ->
                estrofa 
                |> Seq.where (fun verso -> not (String.IsNullOrEmpty(verso)))
                )
|> Seq.mapi (fun i e -> (i+1,e))
|> Seq.iter (fun (i,l) -> printfn "%d: %A" i l)
```

Unlike `ReadAllLines`, the `ReadAllText` function reads the file as a string

```python
let readFileAsTxt (fileName: string) =  
    let lines = File.ReadAllText(fileName)
    lines 
```

```python
let poema = readFileAsTxt fileName
printfn "%A" poema
```

So, we can use the `Split` method of the `System.String` module to separate the text into verses using the `\n` (end of line) character:

```python
let p = poema.Split("\n")
printfn "%A" p
```

Another way to read would be:

```python
let linesOfFile = 
    seq {
        use reader = new StreamReader(File.OpenRead("../data/The Beatles songs dataset.csv"))
        while not reader.EndOfStream do
                yield reader.ReadLine()
    }
```

### Writing to files

We are going to edit the poem a bit so that it can be seen as Markdown.

```python
let poemaConFormato =
    lines
    |> Seq.map (fun l -> l.Replace("tiempo","_tiempo_")) // Cambiamos tiempo por itálica

poemaConFormato
|> Seq.iter (fun v -> printfn "%s" v)
```

```python
File.WriteAllLines("../El reloj de arena.md",poemaConFormato)
```

In order for it to be written as is the poem, we would have to add an empty line between each verse, because a stanza as it is in the file would look like this

> The _time_, since to _time_ and to the destination
They both look alike: the imponderable
Day shadow and the irrevocable course
Of the water that continues on its way.

One way is to insert an empty line between each verse:

```python
let poemaConFormato2 =
    lines
    |> Seq.map (fun l -> l.Replace("tiempo","_tiempo_"))
    |> Seq.map (fun l -> seq {l ; ""})
    |> Seq.concat

poemaConFormato2
|> Seq.iter (fun v -> printfn "%s" v)    
```

Or, following [these instructions](https://www.markdowntutorial.com/lesson/7/), add two spaces to the end of the text.

```python
let poemaConFormato3 =
    lines
    |> Seq.map (fun l -> l.Replace("tiempo","_tiempo_"))
    |> Seq.map (fun l -> l + "  ")

File.WriteAllLines("../data/El reloj de arena 2.md",poemaConFormato3)
```
