## Exercises

### The vending machine, take 32768...

To complete the vending machine model, we need to define the inputs and outputs. Likewise, the manufacturer of the vending machine informs us that there will be at most 7 shelves, and at most 10 products per shelf.
The machine accepts the machine receives coins of 1 doubloon (equivalent to 100 cents), 50, 25 and 10
doubloon cents; or by credit card. Change (if any) is also generated in coins.


#### Extra: an edge case

A rare case, but one that could happen, is the following: the user inserts enough coins
enough to buy the product you want, but the machine does not have enough coins to give
the return







#### Inputs

Using the model I designed in [the coin exercise](https://ibprogramacionavanzada.github.io/fsharp/14_Exercises/#monedas).

- Extend the data types to add the possibility of entering 1 doubloon coins.

In addition, the user must select the product, with a letter (corresponding to the shelf), and a
number (corresponding to the product on the shelf).

- Create the data types to model this input.
- Create a function that validates said input, according to the configuration of the number of shelves and the number of products per shelf.
- Create a function that notifies the client of how much money remains to be entered, in case
the money introduced has not been enough.


#### Outputs

Create the functions that deliver the product to the user (appropriately discounting it from the corresponding stock), and the return, if any.


### Nutrients in the diet

> Do the exercise of producing the code **without looking at the data file!!**

The file [food.csv](https://github.com/fcolavecchia/fp-course-public/blob/main/data/food.csv) extracted from [here](https://data.nal.usda.gov/) contains nutrient data for a large number of feed products. The data correspond in each case to 100 g of each
food product. Each row in the file corresponds to a different product.


- How many categories of nutrients are there?

Building on the previous answer, design the code in such a way that the nutrient data can be accessed _without the need to use indices_.
> Help: `Map...` dictionaries can be your friend.


- Validate the information structure: to do this, check that each of the data rows has the information of all the nutrients
- Find the ten foods with the highest cholesterol content, and the 10 with the lowest cholesterol content.
- Find foods that contain bread (_bread_) and display their carbohydrate content in a table
- Among milk products, find those with the lowest fat content.
