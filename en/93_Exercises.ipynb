### The vending machine app (1/3)

Throughout several exercises we have investigated the software model that simulates a vending machine.
We have seen

- Un [model for coins](https://ibprogramacionavanzada.github.io/fsharp/14_Exercises/) y [Credit cards](https://ibprogramacionavanzada.github.io/fsharp/13_Records_es/)
- Which would be [los inputs y outputs](https://ibprogramacionavanzada.github.io/fsharp/73_Exercises/).
- Cómo [read the products](https://ibprogramacionavanzada.github.io/fsharp/70_VendingMachineReadFood/) y [configure machine](https://github.com/IBProgramacionAvanzada/c3p0/blob/main/TP2/TP2.ipynb).

This exercise proposes to integrate all of this code into a single .NET application. For it

- Design how the code will be organized in modules
- Create the files either using the `dotnet` command or the `Ionide` extension of VS Code.
