### Places in the vending machine

The food vending machine vendor provides the machine with different locations on each shelf to accommodate products that have different sizes. The seller offers three different sizes, single, double and triple.

- Model the sizes of places in a `SlotSize` type.
- Extend the `FoodMachineItem` model to add the size of the product measured by the size of the place it occupies.


### Coffee machine

Our vending machine customer wants to add a coffee machine right next to the food vending machine and wants us to prepare the software to manage it. As a first round conversation, we have been informed that the coffee pot will serve coffee and black tea. Both can be added sugar and cream. You will also have chocolate and cappuccino, with the only option to add sugar. Finally, a healthy green or white tea that can be topped with just milk will also work. Start modeling the coffee machine by entering the types to describe all these drinks.


### Coins

At some point, a customer will enter a certain amount of each currency to purchase a product from the food vending machine. The machine receives 10, 25 and 50 cent coins only.

- Model a `Coins` type that accounts for the number of each of the coins that the machine receives.
- Create a function `amount` that receives a value of `Coins` and returns the amount in $\$$ that the customer entered into the machine.
