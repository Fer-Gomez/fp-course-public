We have received excellent references of your team's skills in functional programming.
using F#, therefore we have decided to give them the job of developing some of the games

The intention is to be able to put the games on the market for the winter holidays, so
**the team must finish their task** by June 29 of this year.

### Deliverables

We need to receive the repository (on GitHub) with the sources of the chosen game, which must contain:

- A .NET solution with the code in F#
- A project with unit tests contained in said solution
- The library projects necessary for the execution of said tests
- A README.md file with information the team deems necessary about the code

The work will be considered delivered with the last _commit_ made in the repository before 11:59 a.m. (GMT-3) on 6/29/2023. Any subsequent _commit_ will not be taken into account.

The team may determine to deliver the proposed work earlier, with a
email communication at [CTO de Terminal Frost](mailto:fcolavecchia@gmail.com).

### Valuation

It will be valued

- Team work
- The general design of the code
- The clarity and readability of the code
- The use of unit tests to test the code
- Aspects of functional programming that are observed in the code

The evaluation will be carried out on the delivered repository, analyzing its evolution and the participation of the team members in the task.
