## Final Test Functional Programming 2023

> This has been proposed as a two or three persons exercise, and originally given two weeks to return it as a github repo. 

Dear Team:

Welcome to [Terminal Frost](https://open.spotify.com/track/4hO2y5DRbeMppklIroS1O8?si=ed84f4fe14ed41e5) Inc.,
a new video game company. Our goal is to develop versions of classic games that can be run on a terminal.
Our clients are programmers who need to relax during their coding, but due to the restrictions of each of the
companies where they work, cannot install games on their computers.

Our first list of games to develop is the following:

- [Tetris](https://en.wikipedia.org/wiki/Tetris)
- [Breakout](https://en.wikipedia.org/wiki/Breakout_(video_game))
- [Pengo](https://en.wikipedia.org/wiki/Pengo_(video_game))
- [Sokoban](https://en.wikipedia.org/wiki/Sokoban)

We have received excellent references of your team's skills in functional programming.
using F#, therefore, we have decided to give you the job of developing some of the games
from the list, at your choice.

The intention is to be able to put the games on the market for the winter holidays, so
**the team must finish their task** in two weeks.

### Deliverables

We need to receive the Github repository with the sources of the chosen game, which must contain:

- A .NET solution with the code in F#
- A project with unit tests contained in said solution
- The library projects necessary for the execution of those tests
- A README.md file with information the team deems necessary about the code

The team may determine which game is going to develop with an 
email communication at [CTO de Terminal Frost](mailto:fcolavecchia@gmail.com).

### Observations

Note the team that:

- The code has to run in the terminal, so if you need to prepare graphics, use letters, numbers and symbols,
For example, a block can be drawn as `[ ]`; the characters, if any, can be described by letters.
- However, we have in our company teams of graphic designers and programmers
specialized in games that will integrate their work into a console-type project that includes the _gameplay loop_.
(for a definition of this concept, you can consult the introduction of [this work](
https://markdangerchen.net/courses/236/guardiola.gameplay%20loop.pdf)). Then, _stick to the requested deliverable_.
