## [The Tennis _Kata_](https://github.com/emilybache/Tennis-Refactoring-Kata)

The purpose of this _kata_ is to model scoring during a tennis match. The score is counted as follows
form:

- A game is won by the first player who has earned at least four total points and at least two points more than the opponent.
- The current score of each game is described in a way peculiar to tennis: scores from zero to three points are described as "zero" (_love_), "15", "30" and "40", respectively.
- If each player has scored at least three points and the scores are equal, the score is "even" (_deuce_).
- If each side has scored at least three points and one player has one more point than his opponent, the score of the game is said to be "advantage" for the leading player.

Write the types and functions to model the counting of score in a tennis match.
