### The _Kata_ of prices in the supermarket

In general, most of the products in the supermarket have a price defined in pesos, for example,
a package of cookies can cost \$123.75.
But there are also products that are sold by weight, like a kilo of apples or oranges.
In addition, some promotions are frequently seen such as:

- If you have two of the same product, 70% discount on the second unit.
- A can of tuna can be on sale, taking three, pay for two.
- Or it may be that for 1000 pesos one can take three packets of weed.

This is a kata for the shower. It is not necessary at this point to implement it as code, but rather
think about the model one would use to describe, for example, how to add the different prices
of the products in the box. some things to think about

- Is it possible to have price values ​​represented by periodic numbers?
- How to handle rounding?
- How to design the tests for the above cases?








### FizzBuzz _Word_

This is a classic programming exercise. Use your usual programming language.

- Write a program that prints the numbers from 1 to 100. But, when a number is a multiple of three, it has to print `Fizz` instead of the number. And to
numbers that are multiples of five, you have to print `Buzz` instead of the number. Furthermore, if the number
is a multiple of three and five, it should print `FizzBuzz`.

For example,

- 9 is divisible by 3, so `Fizz` should be printed.
- 15 is divisible by 3 and 5, so `FizzBuzz` should be printed.
- 25 is divisible by 5, so `Buzz` should be printed.
- 34 is not divisible by either 3 or 5, so 34 must be printed.

The expected result up from 1 to 15 is:
```
1
2
Fizz
4
Buzz
Fizz
7
8
Fizz
Buzz
11
Fizz
13
14
FizzBuzz
```

- Now consider that you want to print the numbers from 1 to any arbitrary number
_N_ greater than 1.

- Following the previous idea, now add the case of replacing the multiples of seven in
the list by `Whizz`. Of course, in case the number is a multiple of 3 and 7, you should print
`Fizz Wizz`, while in the case where it is a multiple of 3, 5 and 7, it should say `FizzBuzzWhizz`, etc.

- What would be the best way to implement the code to make it easy to add an arbitrary case
of multiples in the future?

### An introduction to a _Kata_

A _kata_ is a programming exercise that can be done in an hour or two. In general
It is presented as an exercise with several parts, which must be solved in the order indicated. TO
Unlike a _just_ exercise, there are some rules to follow.

1. In each part of the exercise, you seek to improve the way in which the code was implemented,
compared to the previous part.
2. It is essential to write _tests_. That is, small programs to test the code in
known cases. The final program that solves the exercise must be able to _pass_ all
the written tests.
